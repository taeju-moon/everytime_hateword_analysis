```python
import json
import pandas as pd
file = open("jsoned_everytime.json", "r")
boards = json.load(file, encoding="cp949")
print(boards["흥보게시판"][1][:10])
file.close()
```

```python
#결과 데이터 프레임 생성
initializing_data = list(map(lambda key: [key, boards[key][3]], boards.keys()))
hate_df = pd.DataFrame(initializing_data, columns=['게시판 이름', '게시판 규모']).dropna()
hate_df
```

```python
#게시글에서 형태소 추출
from konlpy.tag import Okt
Okt = Okt()
    
def refine_words(boards):
    tokenized_words_post = {}
    for board in boards:
        print(f"{board}에서 형태소 추출중...")
        posts = boards[board][1]
        temp = []
        for post in posts:
            tokenized = Okt.morphs(post)
            temp.extend(tokenized)
        tokenized_words_post[board] = temp
        temp = []
    return tokenized_words_post
tokenized_words_post = refine_words(boards)
```

```python
#댓글에서 형태소 추출
from konlpy.tag import Okt
Okt = Okt()
    
def refine_words2(boards):
    tokenized_words_comment = {}
    for board in boards:
        print(f"{board}에서 형태소 추출중...")
        posts = boards[board][2]
        temp = []
        for post in posts:
            tokenized = Okt.morphs(post)
            temp.extend(tokenized)
        tokenized_words_comment[board] = temp
        temp = []
    return tokenized_words_comment
tokenized_words_comment = refine_words2(boards)
```

```python
#혐오 단어 사전 불러오기
import pandas as pd

hate_words = pd.read_csv("hate_word_dictionary.csv", encoding="cp949")
hate_words
```

```python
#혐오 단어 사전에 따라 욕설 및 혐오 단어 빈도 분석
hate_word_dict = {
    "게시글":[],
    "댓글":[],
}

hate_word_dict2 = {
    "인사캠 자유게시판":[],
    "자과캠 자유게시판":[],
    "새내기게시판":[],
    "흥보게시판":[],
    "동아리*학회":[],
    "시사*이슈":[],
    "글로벌융합학부게시판":[],
    "영어영문학과게시판":[],
    "비밀게시판":[],
    "헬스게시판":[],
    "애니메이션 게시판":[],
}
def apply_standard(tokenized_words, type_):
    temp = {}
    
    for board in hate_df["게시판 이름"]:
        temp[board] = 0
    
    def add_to_temp(board):
        temp[board] +=1
    
    for board in tokenized_words.keys(): #모든 게시판에 대해서
        result = {}
        print(f"{board}를 작업중입니다...")
        for index, word in enumerate(tokenized_words[board]):
            if word == "II": #구분자라면 continue
                continue
            isFinished = False
            #standard
            for hate_word in hate_words["standard"]: #시작하거나 끝나거나
                if word.startswith(hate_word) or word.endswith(hate_word):
                    isFinished=True
                    add_to_temp(board)
                    hate_word_dict[type_].append(hate_word)
                    hate_word_dict2[board].append(hate_word)
                    break
            #fixed       
            if not isFinished: #아직 찾아내지 못했다면
                for hate_word in hate_words["fixed"].dropna(): #동일하면
                    if word == hate_word:
                        isFinished=True
                        add_to_temp(board)
                        hate_word_dict[type_].append(hate_word)
                        hate_word_dict2[board].append(hate_word)
                        break
                    
            #special
            if not isFinished: #아직 찾아내지 못했다면
                if word in ["성별", "집단", "정당"] and tokenized_words[board][index-1] == "그":
                    add_to_temp(board)
                    isFinished=True
                    hate_word_dict[type_].append("그 "+word)
                    hate_word_dict2[board].append("그 "+word)
                    break
            
            #specail2
            if not isFinished: #아직 찾아내지 못했다면
                if word in ["문학", "수육", "짜장", "카레"] and tokenized_words[board][index-1] == "해병":
                    add_to_temp(board)
                    isFinished = True
                    hate_word_dict[type_].append("해병 "+word)
                    hate_word_dict2[board].append("해병 "+word)
                    break
            
                        
    return temp
                        
                    
posts_result = apply_standard(tokenized_words_post, "게시글") #게시글 빈도분석
```

```python
print(posts_result)
```

```python
comments_result = apply_standard(tokenized_words_comment, "댓글") #댓글 빈도분석
```

```python
print(comments_result)
```

```python
#데이터 프레임 완성
hate_df["게시글"] = list(posts_result.values())
hate_df["댓글"] = list(comments_result.values())
hate_df["전체"] = hate_df["게시글"] + hate_df["댓글"] #전체 열 == 게시글 열 + 댓글 열
hate_df
```

```python
#카테고리 열 추가
hate_df["카테고리"] = ["자유", "자유", "자유", "정보", "정보", "정치", "학과", "학과", "대자연", "관심사", "관심사"]
```
