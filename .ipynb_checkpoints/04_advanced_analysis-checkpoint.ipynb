```python
import json
file = open("jsoned_hate_words.json", "r")
hate_words = json.load(file, encoding="cp949")
hate_words
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
fig, ax = plt.subplots(figsize=(8,5), facecolor="white")

matplotlib.rc('font', family='Malgun Gothic')

x = np.arange(2)
years = ['게시글', '댓글']
values = [len(hate_words["게시글"]), len(hate_words["댓글"])]

plt.title("게시글 - 댓글 혐오 단어 빈도")
plt.bar(x, values, color='y')
plt.xticks(x, years)

plt.show()
```

```python
import pandas as pd
refine_standard = pd.read_csv("refine_hate_word.csv", encoding="cp949")
refine_standard
```

```python
def apply_standard(morphs):
    for i in range(108):
        if refine_standard.iloc[i,0] == morphs:
            return refine_standard.iloc[i,1]
    

def refine_hate_words(hate_words):
    hate_words["게시글"] = list(map(apply_standard, hate_words["게시글"]))
    hate_words["댓글"] = list(map(apply_standard, hate_words["댓글"]))
    return hate_words

hate_words = refine_hate_words(hate_words)
```

```python
from collections import Counter
from wordcloud import WordCloud
posts_count = Counter(hate_words["게시글"])
comments_count = Counter(hate_words["댓글"])
```

```python
import matplotlib.pyplot as plt
wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(posts_count)
plt.figure(figsize=(8,16))
plt.title("게시판 욕설 및 혐오 표현")
plt.imshow(gen)
```

```python
wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(comments_count)
plt.figure(figsize=(8,16))
plt.title("댓글 욕설 및 혐오 표현")
plt.imshow(gen)
```
