```python
import json
file = open("jsoned_hate_words2.json", "r")
hate_words = json.load(file, encoding="cp949")
hate_words
```

```python
import pandas as pd
refine_standard = pd.read_csv("refine_hate_word.csv", encoding="cp949")
refine_standard
```

# 비슷한 단어 통합

```python
def apply_standard(morphs):
    for i in range(108):
        if refine_standard.iloc[i,0] == morphs:
            return refine_standard.iloc[i,1]
    

def refine_hate_words(hate_words):
    for board in hate_words.keys():
        hate_words[board] = list(map(apply_standard, hate_words[board]))
    return hate_words

hate_words = refine_hate_words(hate_words)
```

# 게시판 counting해서 counted_hate_words에 저장

```python
from collections import Counter
from wordcloud import WordCloud

counted_hate_words = hate_words.copy() #딕셔너리 복사
for board in counted_hate_words.keys():
    counted_hate_words[board] = Counter(counted_hate_words[board])
counted_hate_words["인사캠 자유게시판"]
```

# 시발, 존나, 병신 제거

```python
for board in counted_hate_words.keys():
    del counted_hate_words[board]["시발"]
    del counted_hate_words[board]["존나"]
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc('font', family='Malgun Gothic')
for board in counted_hate_words.keys():
    if len(counted_hate_words[board].keys()) < 10:
        continue
    wc = WordCloud(font_path='malgun', width=400, height=400, scale=2.0, max_font_size=250, background_color="white")
    gen = wc.generate_from_frequencies(counted_hate_words[board])
    plt.figure(figsize=(6,10))
    plt.title(f"{board} 욕설 및 혐오 표현", color="white")
    plt.imshow(gen)
```
