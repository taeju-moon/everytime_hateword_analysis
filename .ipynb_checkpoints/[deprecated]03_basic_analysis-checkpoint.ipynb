```python
import pandas as pd

sentiment_df = pd.read_csv("sentiment_dataframe.csv", encoding="cp949")
sentiment_df.iloc[0]
```

```python
sentiment_df.columns[2:]
```

```python
#pie chart 그리기
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)

for i in range(1,11,1):
    ratio = sentiment_df.iloc[i-1][12:]
    labels = sentiment_df.columns[12:]
    plt.pie(ratio, labels=labels, autopct="%.1f%%")
    plt.title(sentiment_df.iloc[i-1][0])
    plt.show()
```

```python
for i in range(1,11,1):
    ratio = sentiment_df.iloc[i-1][2:12]
    labels = sentiment_df.columns[2:12]
    plt.pie(ratio, labels=labels, autopct="%.1f%%")
    plt.title(sentiment_df.iloc[i-1][0])
    plt.show()
```
