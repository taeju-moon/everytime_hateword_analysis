```python
initializing_data = list(map(lambda key: [key, boards[key][3]], boards.keys()))
sentiment_df = pd.DataFrame(initializing_data, columns=['ê²Œì‹œíŒ ì´ë¦„', 'ê²Œì‹œíŒ ê·œëª¨'])
sentiment_df
```

```python
from konlpy.tag import Okt
Okt = Okt()
    
def refine_words(boards):
    tokenized_words_post = {}
    for board in boards:
        print(f"{board}ì—ì„œ ì–´ê°„ ì¶”ì¶œì¤‘...")
        posts = boards[board][1]
        temp = []
        for post in posts:
            tokenized = Okt.morphs(post)
            temp.extend(tokenized)
        tokenized_words_post[board] = temp
        temp = []
    return tokenized_words_post
tokenized_words_post = refine_words(boards)
```

```python
from konlpy.tag import Okt
Okt = Okt()
    
def refine_words2(boards):
    tokenized_words_comment = {}
    for board in boards:
        print(f"{board}ì—ì„œ ì–´ê°„ ì¶”ì¶œì¤‘...")
        posts = boards[board][2]
        temp = []
        for post in posts:
            tokenized = Okt.morphs(post)
            temp.extend(tokenized)
        tokenized_words_comment[board] = temp
        temp = []
    return tokenized_words_comment
tokenized_words_comment = refine_words2(boards)
```

```python
import pandas as pd

standard = pd.read_csv("sentiment_list.csv", encoding="cp949")
standard = standard.dropna()
standard
```

```python
def get_sentiment(words):
    temp = {}
    for word in words:
        for row_index, std in enumerate(standard["ë‹¨ì–´"]):
            if word in std:
                sentiment = standard.iloc[row_index, 2]
                if sentiment in temp:
                    temp[sentiment] +=1
                else:
                    temp[sentiment] = 1
    return temp

def sentiment_analyize(boards, tokenized_words):
    get_dict_of_results = {}
    for board in boards:
        print(f"{board}ë¥¼ ì‘ì—…ì¤‘ì…ë‹ˆë‹¤...")
        words = filter(lambda x: len(x) >1, tokenized_words[board]) #1ìŒì ˆ ì œì™¸
        temp = get_sentiment(words)
        get_dict_of_results[board] = temp
    return get_dict_of_results
            
post_result = sentiment_analyize(boards, tokenized_words_post)
comment_result = sentiment_analyize(boards, tokenized_words_comment)
```

```python
def get_dict(boards_keys, result):
    returning_dict = {}
    for board in boards_keys:
        _sum = sum(result[board].values())
        returning_dict[board] = {}
        for sentiment in result[board]:
            ratio = result[board][sentiment] / _sum * 100
            returning_dict[board][sentiment] = round(ratio,2)
    return returning_dict

sents_ratio1 = get_dict(boards.keys(), post_result)
sents_ratio2 = get_dict(boards.keys(),comment_result)
sents_ratio1["ì¸ì‚¬ìº  ììœ ê²Œì‹œíŒ"]
```

```python
stds = ["ìŠ¬í””", "ë¶„ë…¸", "ë†€ëŒ", "ê¸°ì¨", "í¥ë¯¸", "ì§€ë£¨í•¨", "í†µì¦","í˜ì˜¤", "ì§€ë£¨í•¨", "ì¤‘ì„±", "ê¸°íƒ€"]
```

```python
def fill_dataframe(sents_ratio1, sents_ratio2):
    array_result1 = []
    array_result2 = []
    for std in stds:
        temp = []
        for board in sents_ratio1.keys():
            try:
                temp.append(sents_ratio1[board][std])
            except:
                temp.append(None)
        array_result1.append(temp)
    for std in stds:
        temp = []
        for board in sents_ratio2.keys():
            try:
                temp.append(sents_ratio2[board][std])
            except:
                temp.append(None)
        array_result2.append(temp)
    for index, std in enumerate(stds):
        sentiment_df[f"{std} - ê²Œì‹œê¸€"] = array_result1[index]
    for index, std in enumerate(stds):
        sentiment_df[f"{std} - ëŒ“ê¸€"] = array_result2[index]

fill_dataframe(sents_ratio1, sents_ratio2)
sentiment_df
```
