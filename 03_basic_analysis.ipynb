```python
import pandas as pd
#저장된 자료 불러옴
hate_df = pd.read_csv("hate_df_result.csv", encoding="cp949")
hate_df
```

```python
#multi chart analysis
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
fig, ax = plt.subplots(figsize=(20,10), facecolor="white")

matplotlib.rc('font', family='Malgun Gothic')

data = [
    list(hate_df["게시글"]),
    list(hate_df["댓글"]),
    list(hate_df["게시판 이름"]),
]


index = np.arange(11)

b1 = plt.bar(index+0.00, data[0], width=0.25, color='coral', label='게시글')

b2 = plt.bar(index+0.25, data[1], width=0.25, color='red', label='댓글')

plt.xticks(ticks = index ,labels = data[2], rotation = 'vertical', fontsize=15)
plt.title("게시판 별 욕설 및 혐오 단어 수")
plt.xlabel("게시판")
plt.ylabel("욕설 및 혐오 단어 수")
plt.legend()
plt.show()
plt.savefig("graph.png")
```

```python
#게시판 규모별 분석
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
fig, ax = plt.subplots(figsize=(8,5), facecolor="white")

matplotlib.rc('font', family='Malgun Gothic')

data = [
    list(hate_df.groupby("게시판 규모")["게시글"].mean()),
    list(hate_df.groupby("게시판 규모")["댓글"].mean()),
    ["규모 큼", "규모 작음"],
]


index = np.arange(2)

b1 = plt.bar(index+0.00, data[0], width=0.25, color='coral', label='게시글')

b2 = plt.bar(index+0.25, data[1], width=0.25, color='red', label='댓글')

plt.xticks(ticks = index ,labels = data[2], rotation = 'vertical')
plt.title("게시판 규모별 욕설 및 혐오 단어 수")
plt.xlabel("규모")
plt.ylabel("욕설 및 혐오 단어 수")
plt.legend()
plt.show()

```

```python
#상관분석, pearsonr으로 통계적 유의미성 확인
import seaborn as sns
from scipy.stats import pearsonr
plt.figure(facecolor="white")
sns.regplot(x="게시글", y="댓글", data=hate_df)
plt.title("게시글 - 댓글 혐오 단어 빈도수")
plt.show()
print(pearsonr(hate_df["게시글"], hate_df["댓글"]))
```
